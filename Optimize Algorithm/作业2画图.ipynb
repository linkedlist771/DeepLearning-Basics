```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("iteration_hist.csv")
data
```

```python
algorithm_names = data.columns[1:]
algorithm_names
```

```python
# 定义一个用来处理类似这样的[1 2] 这样的字符串，并返回其中值的函数
def parse_list(li):
    li = li[1:-1]
    x,y = li.split()
    x = eval(x)
    y = eval(y)
    return np.array([x, y ])
```

```python
def plot_learning_curve(x):
    # 输入的是x的列表
    #print(x.shape)
    x = x[1:]
    y  =  x[:,0]**2/20+x[:,1]**2
    
    #print(size)
    x = x[::500]
    y = y[::500]
    size = len(x)
    
    plt.plot(np.arange(size)*500,y,'x--')
```

```python
plt.figure(figsize=(12,8))
plt.style.use('seaborn-whitegrid')
for algorithm in algorithm_names:
    name = algorithm 
    x = data[name].values
    x = np.array([parse_list(i) for i in x])
    plot_learning_curve(x)
    plt.legend(algorithm_names)
plt.xlabel("Iteration")
plt.ylabel("$x^2/20+y^2$")
plt.savefig("迭代学习曲线.png", dpi=300)
plt.show()
```
